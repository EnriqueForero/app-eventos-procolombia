```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse, urljoin

def obtener_urls(sitio_web):
    # Hacemos la solicitud HTTP
    respuesta = requests.get(sitio_web)
    # Verificamos que la solicitud sea exitosa
    if respuesta.status_code == 200:
        # Creamos un objeto BeautifulSoup para analizar el HTML
        soup = BeautifulSoup(respuesta.text, 'html.parser')
        # Obtenemos todos los elementos 'a' (enlaces) del HTML
        enlaces = soup.find_all('a')
        urls = []
        # Iteramos sobre los enlaces
        for enlace in enlaces:
            # Obtenemos el atributo 'href' de cada enlace
            url = enlace.get('href')
            # Si la URL no es nula y no es un enlace interno (comienza con '#')
            if url and not url.startswith('#'):
                # Resolvemos la URL relativa para obtener la URL absoluta
                url_absoluta = urljoin(sitio_web, url)
                # Agregamos la URL a la lista si no está ya presente
                if url_absoluta not in urls:
                    urls.append(url_absoluta)
        return urls
    else:
        # Si la solicitud no es exitosa, imprimimos un mensaje de error
        print("Error al obtener la página:", respuesta.status_code)
        return []

# Ejemplo de uso
sitio_web = "https://www.uniandes.edu.co/"
sitio_web = "https://www.javeriana.edu.co"
urls_encontradas = obtener_urls(sitio_web)
for url in urls_encontradas:
    print(url)
```
